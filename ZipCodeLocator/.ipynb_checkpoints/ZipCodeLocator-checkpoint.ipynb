```python
import numpy as numpy
import pandas as pd
from geopy.geocoders import Nominatim

crimedata = pd.read_excel("savedGeneratedCrimeData.xlsx")
```

```python
def get_zipcode(latitude, longitude):
    geolocator = Nominatim(user_agent="zipcode_finder")
    location = geolocator.reverse((latitude, longitude), addressdetails=True)
    if 'postcode' in location.raw['address']:
        return location.raw['address']['postcode']
    else:
        return "Zipcode not found"
```

```python
latitudeData = crimedata["Latitude"]
longitudeData = crimedata["Longitude"]
crimeType = crimedata["Primary Type"]
zipcodes = crimedata['Zipcode']
crimedata.head(100)
```

```python
for x in range(1000):
    crimedata.at[x, 'Zipcode'] = get_zipcode(latitudeData[x], longitudeData[x])
```

```python
crimedata.to_excel('savedGeneratedCrimeData.xlsx')
```
